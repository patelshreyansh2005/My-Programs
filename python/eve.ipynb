```python
def toBinary(n):
    l = helper(int(n),[])
    s = ""
    for i in l:
        s+=str(i)
    return s

def helper(n,l):
    if(int(n)==0):
        return l
    l = helper(int(n)//2,[])
    l.append(int(n)%2)
    return l

print(toBinary(8))
```
