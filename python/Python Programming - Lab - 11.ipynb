### 01) WAP to demonstrate the use of Pie chart.

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
x=[random.randint(0,i) for i in range(1,10)]
y=[random.randint(0,i) for i in range(1,10)]

plt.plot(x,y,c='b')
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x = [random.randint(1,10) for i in range(1,6)]
y = [random.randint(1,10) for i in range(1,6)]
l = ['1st','2nd','3rd','4th','5th']
c = [random.choice('rbgcmy') for i in range(1,6)]
plt.title('sem wise spi')
plt.bar(x,y,color=c,label=l)
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
cpis = [random.random() for i in range(1,1001)]
plt.hist(cpis,color='m',bins=100,histtype='step',align='mid',label='CPI hist')
plt.legend()
plt.show()
```
